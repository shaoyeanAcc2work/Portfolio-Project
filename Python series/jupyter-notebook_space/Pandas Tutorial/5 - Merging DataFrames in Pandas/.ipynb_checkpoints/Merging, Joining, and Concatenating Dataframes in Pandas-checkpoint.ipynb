```python
df2 = pd.read_csv(r"D:\AlexTheAnalyst\PythonCode-master\jupyter-notebook_space\Pandas Tutorial\5 - Merging DataFrames in Pandas\LOTR 2.csv")
df2
```

```python
df1.merge(df2, how = 'inner', on = 'FellowshipID', suffixes=('_left','_right'))
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID','FirstName'])
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
df1.merge(df2, how = 'cross')
```

```python
df1.join(df2, on = ('FellowshipID'), how='outer', lsuffix='_df1', rsuffix='_df2')
```
